# Estimated checks until detection of a bad piece

## 98% good pieces

If a process generates good pieces 98% of the time, how many consecutive good pieces can you expect to sample before you detect a bad piece?

```r
sim2pct <- diff(which(rbinom(3000000, 1, 0.98) %in% 0))
```

```r
den <- density(sim2pct)
plot(den)
```

```r
print("Median")
print(median(sim2pct))
```

```r
print("Mean")
print(mean(sim2pct))
```

```r
print("Mode")
modeTable <- table(sim2pct)
print(names(modeTable[modeTable==max(modeTable)]))
```
